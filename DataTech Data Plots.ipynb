```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

#include same head and tail
df = pd.read_csv('/Users/ejoe/PycharmProjects/DataTechPlots/MI_vaccine_CLEANED.csv').head(100).tail(100) 
df
```

```python
# PIE CHART - vaccine distributed by type
vaccine_type = df['Vaccine Type'].to_list()
pfizer = []
moderna = []
jj = []

for value in vaccine_type:
    if value == 'Pfizer':
        pfizer.append(value)
    if value == 'Moderna':
        moderna.append(value)
    if value == 'J&J':
        jj.append(value)
        
slices = [len(pfizer), len(moderna), len(jj)]
vaccine = ['Pfizer', 'Moderna', 'Johnson & Johnson']
cols = ['c','m','r']
plt.figure(figsize=(10,10))
plt.pie(slices,
        labels=vaccine,
        colors=cols,
        startangle=True,
        shadow=False,
        explode=(0, 0, 0),
        autopct='%1.1f%%')
plt.title('Michigan Vaccine Type Distributed')
plt.show()
```

```python
female_vaccine = df['Vaccine Type'][df['Sex'] == 'F']
male_vaccine = df['Vaccine Type'][df['Sex'] == 'M']
f_pf = []
f_m = []
f_jj = []
for value in female_vaccine:
    if value == 'Pfizer':
        f_pf.append(value)
    if value == 'Moderna':
        f_m.append(value)
    if value == 'J&J':
        f_jj.append(value)
m_pf = []
m_m = []
m_jj = []
for value in male_vaccine:
    if value == 'Pfizer':
        m_pf.append(value)
    if value == 'Moderna':
        m_m.append(value)
    if value == 'J&J':
        m_jj.append(value)
x = np.arange(3)
y1 = [len(f_pf), len(f_m), len(f_jj)]
y2 = [len(m_pf), len(m_m), len(m_jj)]
width = 0.2
# plot data in grouped manner of bar type
plt.figure(figsize=(20,10))
plt.bar(x - 0.2, y1, width, color='orange')
plt.bar(x, y2, width, color='cyan')
plt.xticks(x, ['Pfizer', 'Moderna', 'Johnson & Johnson'])
plt.xlabel('Vaccine Type')
plt.ylabel('Distribution #')
plt.title('Vaccine Type by Gender')
plt.legend(['Female','Male'])
plt.show()     
plt.stackplot
```
